```python
# Create a 2x4 grid of subplots
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,12))

# Span subplot over two columns
# ax[1,2] = plt.subplot2grid((2,4), (1,2), colspan=2)

for i,col in enumerate(categories[:len(categories)-1]):
    
    # Create a cross tabulation showing the proportion of purchased and non-purchased loans for each category of the feature
    cross_tab = pd.crosstab(index=df[col], columns=df['price_range'])
    
    # Using the normalize=True argument gives us the index-wise proportion of the data
    cross_tab_prop = pd.crosstab(index=df[col], columns=df['price_range'], normalize='index')

    
    # Plot stacked bar charts
    x, y = i//3, i%3
    cross_tab_prop.plot(kind='bar', ax=ax[x,y], stacked=True, width=0.8, legend=False, ylabel='Proportion', sharey=True)
    
    # Add the proportions and counts of the individual bars to our plot
    for idx, val in enumerate([*cross_tab.index.values]):
        for (proportion, count, y_location) in zip(cross_tab_prop.loc[val],cross_tab.loc[val],cross_tab_prop.loc[val].cumsum()):
            ax[x,y].text(x=idx-0.25, y=(y_location-proportion)+(proportion/2)-0.03,
                         s = f'   {count}\n({np.round(proportion * 100, 1)}%)', 
                         color = "black", fontsize=9, fontweight="bold")
    
    # Add legend
    ax[x,y].legend(title='Price Range', loc=(0.24,0.92), fontsize=8, ncol=4)
    # Set y limit
    ax[x,y].set_ylim([0,1.12])
    # Rotate xticks
    ax[x,y].set_xticklabels(ax[x,y].get_xticklabels(), rotation=0)
    # Increase xlabel & ylabel font sizes
    ax[x,y].set_ylabel('Proportion', fontsize=15)
    ax[x,y].set_xlabel(col, fontsize=15)        

plt.suptitle('Categorical Features vs Price Range Stacked Barplots', fontsize=22)
plt.tight_layout()                     
plt.show()
```

# <b><span style='color:#0000FF;'>Step 7.1 |</span><span style='color:#0000FF;'>  Logistic Regression</span></b>

Logistic Regression predicts the probability that an observation belongs to each category of the target variable, with the probabilities summing up to 1 across all categories. It uses softmax function.

<b><span style='color:#0000FF;'>Advantages of Logistic Regression:</span></b>

1. __Handles multiple classes:__ Logistic regression can handle situations where the dependent variable has more than two categories.
2. __Interpretability:__ The model provides interpretable coefficients for each predictor variable, allowing for easy understanding of the impact of each variable on the outcome.
3. __No assumptions of linearity:__ Unlike linear regression, multinomial logistic regression does not assume a linear relationship between the independent variables and the logit of the dependent variable.

<b><span style='color:#0000FF;'>Disdvantages of Logistic Regression:</span></b>

1. __Overfitting:__ Logistic regression can overfit the data if the model complexity is too high relative to the available data.

2. __Sensitivity to outliers:__ Logistic regression can be sensitive to outliers, especially if the dataset is small or if there are extreme values in the predictor variables.

<a id="logistic_init_train"></a>
<b><span style='color:#0000FF;'>Step 7.1.1 |</span><span style='color:#0000FF;'>Logistic Regression: Initiation and training</span></b>

**Pipline** is a series of data processing steps chained together sequentially. Each step in the pipeline typically performs a specific transformation or operation on the data.

<a id="logistic_eval"></a>
<b><span style='color:#0000FF;'>Step 7.1.2 |</span><span style='color:#0000FF;'>Logistic Regression: Evaluation</span></b>

- **Accuracy Score**: In classification problem, __accuracy score__ is used to see the performance. It shows the proportion of correctly predicted classes out of the total classes. If a model correctly predicts 90 out of 100 classes, its accuracy score would be 90%.

- **Classification Report**: In classification problem, __classification report__ shows the concise summary. It provides some key metrics like `precision`, `recall`, `F1-score`, and `support` for each class, as well as the overall accuracy.

    - __Precission__: Precission shows the proportion of true positive predictions out of all positive predictions. It indicates the model's ability to avoid false positives.

__Support Vector Machine (SVM)__ is a popular machine learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that separates the data points into different classes. The hyperplane is determined by maximizing the margin, which is the distance between the hyperplane and the nearest data points of each class.

<b><span style='color:#0000FF;'>Advantages of SVM:</span></b>

1. __Robust to noise & outliers__ - SVM is less sensitive to noise & outliers compared to other algorithms, making it suitable for tasks where the presence of noise & outliers is expected.
2. __Good generalization__ - SVM is known for its ability to generalize well, meaning it can perform well on new and unseen data.
3. __Handles non-linear data__ - SVM can handle non-linear decision boundaries, which makes it useful in situations where the data is not linearly separable.
4. __Flexibility in kernel selection__ - SVM provides flexibility in selecting different kernel functions such as linear, polynomial, and radial basis function (RBF), allowing it to work with a wide range of datasets.
5. __Effective in high-dimensional spaces__ - SVM can work well with high-dimensional datasets, which is useful in situations where the number of features exceeds the number of samples.

<b><span style='color:#0000FF;'>Disdvantages of Logistic Regression:</span></b>

1. __Computationally intensive__ - SVM is computationally intensive, meaning it can be slow and require a lot of memory when working with large datasets.
2. __Sensitivity to hyperparameters__ - SVM is sensitive to the choice of parameters such as the kernel function and regularization parameters, which can affect its performance.
3. __Limited interpretability__ - SVM does not provide explicit models for understanding the relationship between features and the target variable, making it difficult to interpret the results.
4. __Bias-Variance Tradeoff__ - Like all machine learning algorithms, SVM is subject to the bias-variance tradeoff, meaning that it can be overfit or underfit the data, leading to poor performance.

__Decision Tree__ Decision trees recursively split the data into subsets based on the values of input features. At each node of the tree, a decision is made based on a feature's value, leading to the creation of branches. The process continues recursively until a stopping criterion is met, such as reaching a maximum depth, minimum number of samples, or purity of leaf nodes. Leaf nodes represent the output.

<b><span style='color:#0000FF;'>Advantages of Decision Tree:</span></b>

1. Interpretability: Decision trees are easy to interpret and visualize, making them useful for explaining the reasoning behind decisions to non-experts.
2. Handling Non-linearity and Interactions: Decision trees can capture complex non-linear relationships and interactions between features in the data without requiring complex transformations.

3. No Assumptions about Data Distribution: Decision trees make no assumptions about the distribution of the data or the relationships between features, making them robust to outliers and irrelevant features.

<b><span style='color:#0000FF;'>Disdvantages of Decision Tree:</span></b>

1. High Variance: Decision trees are sensitive to small variations in the training data, leading to high variance in the learned models. This sensitivity can result in different trees being produced for slightly different training datasets.

2. Lack of Robustness: Decision trees are sensitive to the structure of the data and can produce different trees for different subsets of the data. Small changes in the data or the addition of new data points can lead to significant changes in the learned model.

1. **Criterion**:
   - The criterion hyperparameter determines the quality measure used for making splits in a decision tree.
   - Common options include "gini" for Gini impurity and "entropy" for information gain.
   - Gini impurity measures the probability of incorrectly classifying a randomly chosen element.
   - Entropy measures the degree of disorder or randomness in the data.
   - Tuning: Select the criterion that results in better performance on your dataset, often through cross-validation.

2. **Splitter**:
   - The splitter hyperparameter determines the strategy used to choose the split at each node.
   - Common options are "best" to choose the best split or "random" to choose the best random split.
   - Tuning: Typically, the "best" splitter performs well in most scenarios, but you may experiment with "random" if dealing with noisy data or large datasets.

3. **Max Depth**:
   - The max_depth hyperparameter sets the maximum depth of the decision tree.
   - A higher value allows the tree to grow deeper, potentially capturing more complex relationships in the data.
   - However, deeper trees may lead to overfitting, especially with noisy data.
   - Tuning: Experiment with different values for max_depth to find the optimal trade-off between model complexity and performance.

4. **Min Samples Split**:
   - The min_samples_split hyperparameter sets the minimum number of samples required to split an internal node.
   - If a node has fewer samples than min_samples_split, it will not be split, effectively controlling the tree's depth and preventing overfitting.
   - Tuning: Adjust min_samples_split to control the model's tendency to overfit. Higher values result in simpler trees and help prevent overfitting.

5. **Min Samples Leaf**:
   - The min_samples_leaf hyperparameter sets the minimum number of samples required to be at a leaf node.
   - Leaf nodes with fewer samples than min_samples_leaf may be pruned, preventing the creation of nodes with very few samples and reducing overfitting.
   - Tuning: Similar to min_samples_split, adjusting min_samples_leaf helps control the model's complexity and tendency to overfit.

<b><span style='color:#0000FF;'>Step 7.4.4 |</span><span style='color:#0000FF;'> Random Forest
: Hyperparameter Tuning and cross validation and Training</span></b>

__Hyperparameter Tuning__: Hyperparameter tuning is the process of finding the optimal hyperparameters for a model.

__Cross Validation__: Cross-validation is a technique used to evaluate the performance of a machine learning model by partitioning the dataset into multiple subsets and iteratively training and evaluating the model on different combinations of these subsets.

Hyperparameter tuning for Random Forest involves adjusting the values of hyperparameters such as n_estimators, criterion, max_depth, min_samples_split, and min_samples_leaf to optimize the model's performance.

1. **n_estimators**:
   - n_estimators determines the number of decision trees to be used in the Random Forest ensemble.
   - Increasing n_estimators can improve the model's performance, as it averages out the predictions of multiple trees, reducing overfitting and increasing robustness.
   - However, increasing n_estimators also increases computational complexity and training time.
   - Tuning: Select the optimal number of estimators based on model performance, balancing accuracy with computational resources.

2. **criterion**:
   - The criterion hyperparameter determines the quality measure used for making splits in each decision tree.
   - Common options include "gini" for Gini impurity and "entropy" for information gain.
   - Gini impurity measures the probability of incorrectly classifying a randomly chosen element.
   - Entropy measures the degree of disorder or randomness in the data.
   - Tuning: Choose the criterion that results in better performance on your dataset, often through cross-validation.

3. **max_depth**:
   - max_depth sets the maximum depth of each decision tree in the Random Forest.
   - A higher value allows the trees to grow deeper, potentially capturing more complex relationships in the data.
   - However, deeper trees may lead to overfitting, especially with noisy data.
   - Tuning: Experiment with different values for max_depth to find the optimal balance between model complexity and performance.

4. **min_samples_split**:
   - min_samples_split sets the minimum number of samples required to split an internal node in each decision tree.
   - If a node has fewer samples than min_samples_split, it will not be split, effectively controlling the tree's depth and preventing overfitting.
   - Tuning: Adjust min_samples_split to control the model's tendency to overfit. Higher values result in simpler trees and help prevent overfitting.

5. **min_samples_leaf**:
   - min_samples_leaf sets the minimum number of samples required to be at a leaf node in each decision tree.
   - Leaf nodes with fewer samples than min_samples_leaf may be pruned, preventing the creation of nodes with very few samples and reducing overfitting.
   - Tuning: Similar to min_samples_split, adjusting min_samples_leaf helps control the model's complexity and tendency to overfit.

```python
rf_params = {
    'randomforestclassifier__n_estimators': range(25, 151, 25),
    'randomforestclassifier__criterion' : ['gini', 'entropy'],
    'randomforestclassifier__max_depth' : [3, 5, 7, 10],
    'randomforestclassifier__min_samples_split' : range(2, 11, 1),
    'randomforestclassifier__min_samples_leaf' : range(2, 10, 1),
}
```

# <b><span style='color:#0000FF;'>Step 7.5 |</span><span style='color:#0000FF;'>  Gradient Boosting</span></b>

__Gradient Boosting__ is a powerful ensemble learning technique that builds a strong predictive model by combining the predictions of multiple weaker models, typically decision trees.

<b><span style='color:#0000FF;'>Advantages of Gradient Boosting:</span></b>

1. High Predictive Accuracy: Gradient Boosting often produces highly accurate models, surpassing many other machine learning algorithms, including Random Forests and traditional decision trees.
2. Handles Heterogeneous Data: Gradient Boosting can handle a variety of data types (e.g., numerical, categorical) and can work with missing values without requiring preprocessing.


<b><span style='color:#0000FF;'>Disdvantages of Gradient Boosting:</span></b>

1. Computationally Expensive: Training a Gradient Boosting model can be computationally expensive, especially for large datasets and when using a large number of trees or complex models.
2. Sensitive to Hyperparameters: Gradient Boosting requires careful tuning of hyperparameters, such as the learning rate, number of trees, and tree depth, to achieve optimal performance.
